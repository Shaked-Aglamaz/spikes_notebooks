```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dir = r"D:\Shaked_data\test1"
csv_path = f"{dir}/scores.csv"
df = pd.read_csv(csv_path)

df_filtered = df[df["fold"] == "mean"].drop(columns=["sub"])

# Melt the dataframe to long format for seaborn
df_melted = df_filtered.melt(id_vars=["fold"], var_name="Metric", value_name="Score")

# Create the boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_melted, x="Metric", y="Score", hue="Metric", dodge=False, palette="Set2")
sns.stripplot(data=df_melted, x="Metric", y="Score", dodge=False, jitter=False, alpha=0.7, marker="o", size=6)
plt.legend([],[], frameon=False)  # Hide duplicate legend
plt.xticks(rotation=45)
plt.title("Box Plot of Metrics with Subject Scores")
plt.show()
```
